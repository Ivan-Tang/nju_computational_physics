## 随机游走问题

```python
import numpy as np  #加载数值计算函数库numpy，赋予别名np
import matplotlib.pylab as pl  #加载图形函数库matplotlib中的pylab模块，赋予别名pl
import random  #加载随机数函数库

nstep = 10000
    
xlist = []  # 创建新列表，存储随机游走轨迹(x,y)
ylist = []   
x = 0.0   #随机游走初始位置
y = 0.0 
xlist.append(x) # 将随机游走初始位置添加到轨迹列表
ylist.append(y) 

random.seed(None) #随机数种子
for i in range(0,nstep): # range(n0,step) 产生0到nstep-1的整数序列。 
  theta = random.random()*2.0*3.1415927 #生成（0，2pi）随机游走方向
  dx = np.cos(theta)  #x方向游走距离
  dy = np.sin(theta)
  x = x + dx #更新位置
  y = y + dy         
  xlist.append(x) #追加位置当列表
  ylist.append(y) #追加位置当列表

pl.plot(xlist, ylist,'k-',lw=1) #画出轨迹
pl.show()
```
