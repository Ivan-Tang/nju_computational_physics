```python
import random
random.seed(42)
nsetp = 10000
xlist = []
ylist = []
d_list = []

(x, y) = (0, 0)


for i in range(nsetp):
    xlist.append(x)
    ylist.append(y)
    d_list.append(np.sqrt(x**2 + y**2))
    theta = random.random() * 2 * np.pi
    (x, y) = (x + np.cos(theta), y + np.sin(theta))

#
open('random_walk.dat', 'w').close()
with open('random_walk.dat', 'a') as f:
    for i in range(nsetp):
        f.write(f"{xlist[i]} {ylist[i]} {d_list[i]}\n")
print("Data saved to random_walk.dat")
        
plt.plot(xlist, ylist)
plt.show()
```

```python
nstep = 1000
nwalker = 1000

r2 = np.zeros(nstep, dtype = float)

for iwalker in range(nwalker):
    random.seed(iwalker)
    (x, y) = (0, 0)
    for istep in range(nstep):
        theta = random.random() * 2 * np.pi
        (x, y) = (x + np.cos(theta), y + np.sin(theta))
        r2[istep] += x**2 + y**2

r2 /= nwalker

plt.plot(r2)
plt.xlabel("Time step")
plt.ylabel("Mean square distance")
plt.show()
```
