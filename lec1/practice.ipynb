```python
import numpy as np
import random
import math
import matplotlib.pyplot as plt

x_list = []
y_list = []
r_list = []
r0 = 1
nstep =  10000
x, y = 0, 0

for i in range(nstep):
    theta = random.uniform(0, 2*math.pi)
    dx = r0*np.cos(theta)
    dy = r0*np.sin(theta)
    x += dx
    y += dy
    r = np.sqrt(x**2 + y**2)

    x_list.append(x)
    y_list.append(y)
    r_list.append(r)

#计算方差
plt.plot(np.arange(nstep), r_list)
```
