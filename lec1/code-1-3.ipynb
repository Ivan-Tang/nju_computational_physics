```python
import matplotlib.pylab as pl  #加载图形函数库matplotlib中的pylab模块，赋予别名pl

istep = []
xlist = []
ylist = []

with open('walk-traj.dat', 'r')as infile:
    for lines in infile:
        words = lines.split()
        istep.append(int(words[0]))
        xlist.append(float(words[1]))
        ylist.append(float(words[2]))

pl.plot(xlist, ylist)
pl.show()
```
