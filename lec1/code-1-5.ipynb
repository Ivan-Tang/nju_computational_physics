## 随机游走问题 

```python
import numpy as np  #加载数值计算函数库numpy，赋予别名np
import matplotlib.pylab as pl  #加载图形函数库matplotlib中的pylab模块，赋予别名pl
import random  #加载随机数函数库

nstep = 100  # 把整型数据10000赋予变量 nstep
nwalker = 10000
r2_ave = np.zeros(nstep, dtype=float)
ree = np.zeros(nwalker, dtype=float)
for iwalker in range(0,nwalker):
   x = 0.0   #随机游走初始位置
   y = 0.0 
   random.seed(None) #随机数种子
   for i in range(0,nstep): # range(n0,step) 产生0到nstep-1的整数序列。 
      theta = random.random()*2.0*3.1415927 #生成（0，2pi）随机游走方向
      dx = np.cos(theta)  #x方向游走距离
      dy = np.sin(theta)
      x = x + dx #更新位置
      y = y + dy 
      r2 = x**2 + y**2
      r2_ave[i] = r2_ave[i] + r2/nwalker
   ree[iwalker] = np.sqrt(r2)

hist, edge = np.histogram(ree,bins=50)
dist = 0.5*(edge[0:-1]+edge[1:])
prob = hist/(2*3.1415927*dist)

pl.plot(dist,prob)
pl.xlabel('ree')
pl.ylabel('prob density')
pl.show()

```
