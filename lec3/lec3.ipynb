```python
#欧拉法
import numpy as np 
import matplotlib.pyplot as plt  

R = 2
C = 1
E = 1
t = 0
dt =  0.1
q = 1
n = 100
q_t = []
q_analytical = []
t_range = np.arange(1, 101, 1)

for i in range(n): 
    q = q - q/(R*C) * dt
    t = t + dt
    q_t.append(q)
    q_ana = np.exp(-t/(R*C))
    q_analytical.append(q_ana)
plt.plot(t_range, q_t, 'ro')
plt.plot(t_range, q_analytical)
plt.show()

```

```python
#改进欧拉法
t = 0 
q_t = []
q_analytical = []
q = 1
for i in range(n):
    q_t.append(q)
    q1 = q - q/(R*C) * dt
    q = q - 1/2 * (q+q)/(R*C) * dt
    t += dt
    #q_t.append(q)
    q_analytical.append(np.exp(-t/(R*C)))
plt.plot(t_range, q_t, 'ro')
plt.plot(t_range, q_analytical)
plt.show()

```

```python
#二阶龙格-库塔法
t = 0 
q_t = []
q_analytical = []
q = 1
for i in range(n):
    q1 = q - q/(R*C) * dt * 0.5
    q = q - q1/(R*C) * dt
    t += dt
    q_t.append(q)
    q_ana = np.exp(-t/(R*C))
    q_analytical.append(q_ana)
plt.plot(t_range, q_t, 'ro')
plt.plot(t_range, q_analytical)
plt.show()
```

```python
#四阶龙格-库塔
t = 0 
q_t = []
q_analytical = []
q = 1
for i in range(n):
    k1 = -q/(R*C)
    k2 = -(q+k1*dt/2)/(R*C)
    k3 = -(q+k2*dt/2)/(R*C)
    k4 = -(q+k3*dt)/(R*C)
    q = q + (k1 + 2*k2 + 2*k3 + k4) * dt/6
    t += dt
    q_t.append(q)
    q_ana = np.exp(-t/(R*C))
    q_analytical.append(q_ana)
plt.plot(t_range, q_t, 'ro')
plt.plot(t_range, q_analytical)
plt.show()
```
