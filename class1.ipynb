```python
import numpy as np
import matplotlib.pyplot as plt 

x = np.arange(0, 10, 0.1)
plt.plot(x, np.sin(x))
plt.show()
```

```python
import random
random.seed(42)
nsetp = 10000
xlist = []
ylist = []
d_list = []

(x, y) = (0, 0)


for i in range(nsetp):
    xlist.append(x)
    ylist.append(y)
    d_list.append(np.sqrt(x**2 + y**2))
    theta = random.random() * 2 * np.pi
    (x, y) = (x + np.cos(theta), y + np.sin(theta))

plt.plot(xlist, ylist)
plt.show()
```

```python
plt.plot(d_list)
plt.show()
```

```python
from matplotlib.animation import FuncAnimation

t = np.arange(0, 10, 0.01)
x1 = 1*np.sin(3*np.pi*t)
x2 = 0.5*np.sin(8*np.pi*t)
plt.plot(x1, x2)
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()


```

```python
%matplotlib qt5
 
t = np.arange(0, 10, 0.001)
len = 100
plt.ion()
 
for i in range(len):
    phi_range = np.linspace(0, 2*np.pi, len) 
    phi = phi_range[i]  
    x1 = 1*np.sin(3*np.pi*t)
    x2 = 0.5*np.sin(8*np.pi*t+phi)
    plt.clf()  
    plt.plot(x1, x2)
    plt.pause(0.05)  
    plt.ioff()
```
