```python
# -*- coding: utf-8 -*-
from numpy import arange, zeros, linalg
import pylab as pl

n=1000
a = zeros([n,n])
for i in range(n): #构建连接矩阵
    for j in range(n):
        if(i == j + 1):
            a[i,j] = -1
        if(j == i + 1):
            a[i,j] = -1
        if(i == j):
            a[i,j] = 2
eigenValues,eigenVectors = linalg.eigh(a) #计算本征值，本征向量

index=arange(n)
fig = pl.figure(figsize=(12,4))
ax1 =fig.add_subplot(1,2,1)
ax2 =fig.add_subplot(1,2,2)

ax1.plot(index,eigenValues,'ro',label='mode') #输出本征值
ax2.plot(index,eigenVectors.T[6,:],'r-',label='mode') #输出指定本征向量

ax1.set_xlabel(r'mode id', fontsize=20)
ax1.set_ylabel(r'frequence', fontsize=20)
ax2.set_xlabel(r'position', fontsize=20)
ax2.set_ylabel(r'amplitude', fontsize=20)

pl.subplots_adjust(left=0.15,bottom=0.1,top=0.9,right=0.95, \
                   hspace=0.25,wspace=0.3)
pl.show()
```
