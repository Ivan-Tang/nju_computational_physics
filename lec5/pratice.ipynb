```python
from numpy import array, empty
A = array([[2.0, 0.0, 4.0],
          [2.0, 1.0, 0.0],
          [1.0, -1.0, -1.0]])
b = array([1.0, 1.0, 0.0])
N = len(b)

for i in range(N):
    div = A[i, i]
    A[i, :] /= div
    b[i] /= div

    for j in range(i+1, N):
        mult = -A[j, i]
        A[j, :] += mult * A[i, :]
        b[j] += mult * b[i]

x = empty(N)
for m in range(N-1, -1, -1):
    x[m] = b[m]
    for k in range(m+1, N):
        x[m] -= A[m, k] * x[k]

print(x)
```

```python
#LU分解法
from numpy import array, linalg
A = array([[2.0, 0.0, 4.0],
           [2.0, 1.0, 0.0], 
           [1.0, -1.0, -1.0]])
b = array([1.0, 1.0, 0.0])

x = linalg.solve(A, b)
print(x)
```

```python
import numpy as np 
import matplotlib.pyplot as plt

n = 1000
a = np.zeros([n, n])
for i in range(n):
    for j in range(n):
        if i == j + 1 or i == j - 1:
            a[i, j] = -1
        elif i == j:
            a[i, j] = 2
        else:
            a[i, j] = 0

eigenValues, eigenVectors = np.linalg.eigh(a)
idx = eigenValues.argsort()
eigenValues = eigenValues[idx]
eigenVectors = eigenVectors[:, idx]

index = np.arange(n)
plt.plot(index, eigenValues)
```

```python
import numpy as np
import matplotlib.pyplot as plt
n = 900

a = np.zeros([n, n])

for i in range(n):
    for j in range(n):
        if i == j - 1 or i == j + 1:
            a[i, j] = -1
        elif i == j:
            a[i, j] =  4
        else:
            a[i, j] =  0

eigenValues, eigenVectors = np.linalg.eigh(a)
idx = eigenValues.argsort()
eigenValues = eigenValues[idx]
eigenVectors = eigenVectors[:, idx]

index = np.arange(n)
plt.plot(index, eigenValues)
plt.show()
```
