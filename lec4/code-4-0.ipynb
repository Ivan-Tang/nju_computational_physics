```python
import numpy as np
import matplotlib.pylab as pl

N=100
h = 0.01
eps=1e-4
V=0.05
eps0 = 1.0

phi=np.zeros([N+1,N+1],float)
phi[100,:]=V
phinew=np.empty([N+1,N+1],float)

rho=np.zeros([N+1,N+1],float)
rho[20:40,20:40]=-10.0
rho[60:80,60:80]=10.0

delta = 1.0
while delta> eps:
    for i in range(N+1):
        for j in range(N+1):
            if i==0 or i==N or j==0 or j==N:
                phinew[i,j]=phi[i,j]
            else:
                phinew[i,j]=(phi[i+1,j]+phi[i-1,j]+phi[i,j+1]+phi[i,j-1])/4 \
                +h**2*rho[i,j]/(4*eps0)
    delta = np.max(abs(phi-phinew))
    phi,phinew = phinew,phi
pl.imshow(phi,origin='lower')
pl.gray()
pl.show()
```
