```python
# -*- coding: utf-8 -*-
#铁瓷顺磁相变
import numpy as np
import pylab as pl

def f(x,t):
    return x - np.tanh(x/t)

eps = 0.000001
temp = []
mag = []
for t in np.arange(0.0001, 1.0, 0.0001):
    xa = 0.000001
    xb = 2.0
    N = 1000

    for i in range(0, N):
        x = (xa + xb)/2
        fa = f(xa,t)
        fb = f(xb,t)
        fx = f(x,t)
        if (fa*fx > 0.):
            xa = x
        else:
            xb = x

        if (abs(fx) < eps):
            temp.append(t)
            mag.append(x)
            break
        if (i == N-1):
            print ("\n root not found after %d iterations" %N)

pl.plot(temp,mag,'r-',label='mag vs. temp')
pl.xlabel('temp')
pl.ylabel('magnitization')
pl.legend(loc='lower left')
pl.show()
```
