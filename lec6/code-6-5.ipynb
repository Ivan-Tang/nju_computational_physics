```python
# -*- coding: utf-8 -*-
# 铁磁顺磁相变超越方程函数
import numpy as np
import pylab as pl
def f(x,t):
    return x - np.tanh(x/t)
t = 0.5
mt=[]
fm=[]
for m in np.arange(0.0, 2.5, 0.01):
    mt.append(m)
    fm.append(f(m,t))

pl.plot(mt,fm,'r-',label='fm-m')
pl.xlabel('m')
pl.ylabel('f(m)')
pl.legend(loc='upper left')
pl.show()
```
