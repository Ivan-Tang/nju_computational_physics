```python
# -*- coding: utf-8 -*-
#
import numpy as np
import pylab as pl

F=np.zeros([100,100])
for i in np.arange(0,100): #
    for j in np.arange(0,100):
        x0 = i*0.2-10
        y0 = j*0.2-10
        F[i][j] = x0**2+10*y0**2+4*x0*y0
x=10.234  #
y=11.537  #
deltar=0.05  #
fold = x**2+10*y**2+4*x*y #
xx=[]
yy=[]
xx.append(x)
yy.append(y)
for i in np.arange(0,1000):
    dfdx = -2*x-4*y  #
    dfdy = -20*y-4*x #
    deltax = deltar * dfdx # x
    deltay = deltar * dfdy # y   
    x = x + deltax  #x
    y = y + deltay  #y
    xx.append(x)
    yy.append(y)
    if abs(deltax)<1.0e-6 and abs(deltay)<1.0e-6:
       print (x,y)
       break
   
extent = [-12, 12, -12, 12]
fig = pl.figure(figsize=(10,4))
# contour line
ax1 =fig.add_subplot(1,2,1)
levels = np.arange(0.0,2000.0,50.0)
cs = ax1.contour(F,levels,origin='lower',linewidths=2,extent=extent)
ax1.clabel(cs)
ax1.plot(xx, yy, 'r-',linewidth=1.0)
ax1.set_ylabel(r'Y', fontsize=20)
ax1.set_xlabel(r'X', fontsize=20)
ax1.set_xlim(-1,12)
ax1.set_ylim(-12,12)
pl.show()
```
