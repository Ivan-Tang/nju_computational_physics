```python
# -*- coding: utf-8 -*-
# 最陡下降是势能
import numpy as np
import pylab as pl

F=np.zeros([100,100])
for i in np.arange(0,100):
    for j in np.arange(0,100):
        x = i*0.2-10
        y = j*0.2-10
        F[i][j] = x**2+10*y**2+4*x*y

extent = [-10, 10, -10, 10]

fig = pl.figure(figsize=(10,4))
# contour line
ax1 =fig.add_subplot(1,2,1)
levels = np.arange(0.0,1000.0,50.0)
cs = ax1.contour(F,levels,origin='lower',linewidths=2,extent=extent)
ax1.clabel(cs)
ax1.set_ylabel(r'Y', fontsize=20)
ax1.set_xlabel(r'X', fontsize=20)
ax1.set_xlim(-10,10)
ax1.set_ylim(-10,10)
pl.show()
```
